```python
import pandas as pd

df = pd.read_csv('Tweets.csv')
print("Shape:", df.shape)
print("\nSentiment counts:")
print(df['airline_sentiment'].value_counts())
```

```python
df = df[['text', 'airline_sentiment']]
print(df.head())
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_tweet(text):
    text = text.lower()                           # STEP 1: lowercase
    text = re.sub(r'http\S+|www\S+', '', text)   # STEP 2: remove URLs
    text = re.sub(r'@\w+', '', text)              # STEP 3: remove @mentions
    text = re.sub(r'#\w+', '', text)              # STEP 4: remove hashtags
    text = re.sub(r'[^a-z\s]', '', text)          # STEP 5: remove punctuation
    tokens = text.split()
    tokens = [w for w in tokens if w not in stop_words]   # STEP 6: stopwords
    tokens = [lemmatizer.lemmatize(w) for w in tokens]    # STEP 7: lemmatize
    return ' '.join(tokens)

# Test it on the first tweet before applying to all
print("BEFORE:", df['text'][0])
print("AFTER: ", clean_tweet(df['text'][0]))
```

```python
df['clean_text'] = df['text'].apply(clean_tweet)

# Compare before vs after for first 5 tweets
for i in range(5):
    print(f"BEFORE: {df['text'][i]}")
    print(f"AFTER:  {df['clean_text'][i]}")
    print()
```

```python
df.to_csv('cleaned_tweets.csv', index=False)
print("Saved successfully!")
print("Shape:", df.shape)
print("\nSample of cleaned data:")
print(df[['clean_text', 'airline_sentiment']].head())
```
